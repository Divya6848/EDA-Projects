```python
#List of Brandish weapon by crime description
data[data['Crm Cd Desc'].str.contains('BRANDISH WEAPON')]
```

```python
#Average of Victim Age
data = data['Vict Age'].mean()
print(data)
```

```python
#count of each Victim's gender 
data['Vict Sex'].value_counts()
```

```python
#List of Crimes on a particular date
data[data['DATE OCC'] == '2023-12-14']
```

```python
#Victim Age Distribution
plt.figure(figsize=(8, 4), facecolor='lightgreen')
sns.histplot(data['Vict Age'],kde=False, color='skyblue')
plt.title('Victim Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# Victim Gender Distribution
plt.figure(figsize=(6, 4),facecolor='lightgreen')
sns.countplot(data=data, x='Vict Sex')
plt.title('Victim Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
#Distribution of Crime Status solved vs unsolved
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Status Desc', color='lightgreen')
plt.title('Distribution of Crime Status')
plt.xlabel('Status Description')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
#Area which has high crime Frequency
plt.figure(figsize=(10, 6),facecolor='lightpink')
sns.countplot(data=data, y='AREA NAME', color='lightgreen', order=data['AREA NAME'].value_counts().index)
plt.title('Crime Frequency by Area')
plt.xlabel('Number of Crimes')
plt.ylabel('Area Name')
plt.show()
```

```python
# Top Crime Types
plt.figure(figsize=(14,4),facecolor='lightpink')
sns.countplot(data=data, y='Crm Cd Desc',color='lightgreen', order=data['Crm Cd Desc'].value_counts().index[:5])
plt.title('Distribution of Top 5 Crime Types')
plt.xlabel('Number of Crimes')
plt.ylabel('Crime Description')
plt.show()
```

```python
#Top 5 Weapons Used in Crimes
plt.figure(figsize=(14, 6),facecolor='lightpink')
sns.countplot(data=data, y='Weapon Desc', order=data['Weapon Desc'].value_counts().index[:5])
plt.title('Top 5 Weapons Used in Crimes')
plt.xlabel('Number of Crimes')
plt.ylabel('Weapon Description')
plt.show()
```
