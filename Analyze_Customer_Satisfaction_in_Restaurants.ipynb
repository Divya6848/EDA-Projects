```python
#Rename the foodratings according to their rating points
def food(rating):
    if rating >= 5.0:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return"Avearge"

data['FoodRating']= data['FoodRating'].apply(food)
data.head()
```

```python
#Distribution of HighSatisfaction
plt.figure(figsize=(6, 4))
sns.countplot(x='HighSatisfaction', data=data)
plt.title('Distribution of HighSatisfaction')
plt.show()
```

```python
#Distribution the count of Timeofvisit by DiningOccasion
plt.figure(figsize=(6, 4))
sns.histplot(data=data, x="TimeOfVisit", hue="DiningOccasion", multiple="dodge", shrink=.8)
plt.show()
```

```python
#Distribution of High ratings
fig, a = plt.subplots(1, 3, figsize=(16, 5))
sns.histplot(data['ServiceRating'], bins=5, kde=True, ax=a[0])
a[0].set_title('Distribution of Service Rating')
sns.histplot(data['FoodRating'], bins=5, kde=True, ax=a[1])
a[1].set_title('Distribution of Food Rating')
sns.histplot(data['AmbianceRating'], bins=5, kde=True, ax=a[2])
a[2].set_title('Distribution of Ambiance Rating')
plt.show()
```

```python
#Distribution of Preferred Cuisines
plt.figure(figsize=(6, 6))
cuisine = data['PreferredCuisine'].value_counts()
plt.pie(cuisine, labels=cuisine.index, autopct='%1.1f%%')
plt.title('Distribution of Preferred Cuisines')
plt.show()
```
