```python
#total number of bike rentals
print(data['cnt'].sum())
```

```python
#most popular season for bike rentals
season_labels = {1:"Winter", 2:"Spring", 3:"Summer", 4:"Fall"}
data["season"] = data["season"].map(season_labels)
print(data.groupby('season')['cnt'].sum().idxmax())
```

```python
#Bike Rental distribution across Week
plt.figure(figsize=(8, 5),facecolor="lightyellow")
plt.suptitle('Bike Rental distribution across Week')
sns.lineplot(x='weekday', y='cnt', data=data, marker='o')
plt.show()
```

```python
#Bike Rental distribution Across month
plt.figure(figsize=(10, 5),facecolor="lightyellow")
plt.suptitle('Bike Rental distribution Across month')
sns.lineplot(x='mnth', y='cnt', data=data, marker='o')
plt.show()
```

```python
#Distribution of Rental Count
plt.figure(figsize=(10, 6),facecolor="lightyellow")
sns.histplot(data['cnt'], kde=True, color= 'grey')
plt.title('Distribution of Rental Counts')
plt.show()
```

```python
#Distribution of Bike Rental by Season
plt.figure(figsize=(10, 4),facecolor="lightyellow")
plt.suptitle('Countplot of Bike Rental by Season')
season_labels = {1:"Winter", 2:"Spring", 3:"Summer", 4:"Fall"}
data["season"] = data["season"].map(season_labels)
sns.countplot(data=data, x='season', color='Skyblue')
```

```python
#Distribution of temp,atemp,humidiyt,windspeed
fig, a = plt.subplots(2,2, figsize = (10, 8))
sns.histplot(data['temp'], kde=True, ax=a[0,0])
a[0,0].set_title('Histplot of Temp')
sns.histplot(data['atemp'], kde=True, ax=a[0,1])
a[0,1].set_title('Histplot of atemp')
sns.histplot(data['hum'], kde=True, ax=a[1,0])
a[1,0].set_title('Histplot of humidity')
sns.histplot(data['windspeed'], kde=True, ax=a[1,1])
a[1,1].set_title('Histplot of Windspeed')
plt.show()
```

```python
#Bike rental percent of the year
a = data.groupby('yr')['cnt'].sum()
plt.pie(a, labels=['2011', '2012'], autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.title('Bike rental percent of the year')
plt.show()
```

```python
#Avg Bike rental counts across yr,month,holiday,weekday,workingday,weather
fig, a = plt.subplots(2,3, figsize = (20, 10))

sns.boxplot(x='yr', y='cnt', data=data, ax=a[0,0])
a[0,0].set_title('Avg Bike rentals across season')

sns.boxplot(x='mnth', y='cnt', data=data, ax=a[0,1])
a[0,1].set_title('Avg Bike rentals across month')

sns.boxplot(x='holiday', y='cnt', data=data, ax=a[0,2])
a[0,2].set_title('Avg bike rentals across holiday')

sns.boxplot(x='weekday', y='cnt', data=data, ax=a[1,0])
a[1,0].set_title('Avg bike rentals across Weekday')

sns.boxplot(x='workingday', y='cnt', data=data, ax=a[1,1])
a[1,1].set_title('Avg bike rentals across Workingday')

sns.boxplot(x='weathersit', y='cnt', data=data, ax=a[1,2])
a[1,2].set_title('Avg bike rentals across Weather')
plt.show()
```
